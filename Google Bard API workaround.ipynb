```python
from bardapi import BardCookies

#fill the variables with the related values

Secure_1PSID=''
Secure_1PSIDTS=''
Secure_1PSIDCC=''

cookie_dict = {
    "__Secure-1PSID": Secure_1PSID,
    "__Secure-1PSIDTS": Secure_1PSIDTS,
    "__Secure-1PSIDCC": Secure_1PSIDCC
}

bard = BardCookies(cookie_dict=cookie_dict,timeout=300)

#open the file
image = open('topology.PNG', 'rb').read()

#ask question about the image
bard_output=bard.ask_about_image('''
explain the network topology in details by text including all info such as ip addres, loopbacks, vlans 
and how devices are connected to each other via which ports.''', image)

#print the result
print(bard_output['content'])
```

```python
from bardapi import BardCookies

Secure_1PSID=''
Secure_1PSIDTS=''
Secure_1PSIDCC=''

cookie_dict = {
    "__Secure-1PSID": Secure_1PSID,
    "__Secure-1PSIDTS": Secure_1PSIDTS,
    "__Secure-1PSIDCC": Secure_1PSIDCC
}

query='what is the capital of Turkey?'
bard = BardCookies(cookie_dict=cookie_dict)
print(bard.get_answer(query)['content'])
```

sample output:

Ankara has been the capital of Turkey since the founding of the Republic of Turkey in 1923. It is home to the Turkish government, parliament, and presidency. Ankara is also a major economic and cultural center for Turkey.
